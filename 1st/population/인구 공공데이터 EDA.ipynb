```python
import csv
```

```python
file = open('./age.csv')
data = csv.reader(file)
result = []

for row in data :
    if '신도림' in row[0] :
        for i in row[3:] :
            result.append(int(i)) #데이터를 시각화하려면 각 값을 정수형으로 꼭 변경해야함!

print(result)
```

```python
# https://matplotlib.org/stable/api/index.html
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.plot(result)
plt.show()
```

```python
plt.bar(range(101),result)
plt.show()
```

```python
file2 = open('./gender.csv')
data = csv.reader(file2)
m = []
f = []
for row in data :
    if '신도림' in row[0] :
        for i in row[3:104] :
            m.append(-int(i)) ## 반대로 표시하기 위해 -부호 추가
        for i in row[106:] :
            f.append(int(i))
```

```python
# https://matplotlib.org/stable/api/index.html
import matplotlib.pyplot as plt2
plt2.rc('font', family='Malgun Gothic')
plt2.title('신도림 지역의 남녀 성별 인구 분포')
plt2.barh(range(101),m,label='남성')
plt2.barh(range(101),f,label='여성')
plt2.legend()
plt2.show()
```
